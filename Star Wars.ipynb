```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

star_wars = pd.read_csv(r'C:\Users\chizj\Downloads\DataQuest CSVs\starwars.txt', encoding="ISO-8859-1")
```

```python
star_wars.head(10)
```

```python
star_wars.columns
```

```python
# Removing rows where 'RespondentId' is null

star_wars = star_wars[pd.notnull(star_wars['RespondentID'])]
```

```python
star_wars.head()
```

I'm going to explore the next two columns, 

* 'Have you seen any of the 6 films in the Star Wars franchise?'
* 'Do you consider yourself to be a fan of the Star Wars film franchise?'

and I will turn both into boolean values - Yes:True, No:False, NaN:NaN.

```python
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts()
```

```python
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts()
```

```python
yes_no = {'Yes':True, 'No':False}

for col in ['Have you seen any of the 6 films in the Star Wars franchise?',
       'Do you consider yourself to be a fan of the Star Wars film franchise?']:
    star_wars[col] = star_wars[col].map(yes_no)
```

```python
star_wars.head()
```

As you can see in the above, both columns are no booleans.

Now I'll update the columns that are asking if a person has seen a specific Star Wars movie. These would be columns 4-9. I will change these answers to a boolean to quickly check if the person has seen each movie or not.

```python
seen_movie = {'Star Wars: Episode I  The Phantom Menace':True, 'Star Wars: Episode II  Attack of the Clones':True,
             'Star Wars: Episode III  Revenge of the Sith':True, 'Star Wars: Episode IV  A New Hope':True,
             'Star Wars: Episode V The Empire Strikes Back':True, 'Star Wars: Episode VI Return of the Jedi':True,
             np.NaN:False}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(seen_movie)
```

```python
star_wars.head()
```

Now I will rename columns 4-9 so that they are easier to understand.

```python
star_wars = star_wars.rename(columns={'Which of the following Star Wars films have you seen? Please select all that apply.':'seen_1',
                                      'Unnamed: 4':'seen_2', 'Unnamed: 5':'seen_3', 'Unnamed: 6':'seen_4', 'Unnamed: 7':'seen_5',
                                      'Unnamed: 8':'seen_6'})
```

```python
star_wars.head()
```

I am going to make each of the columns 10-15 a numeric data type (float), as well as rename the columns so they are easier to quickly understand.

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
star_wars = star_wars.rename(columns = {'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.':'ranking_1',
                                        'Unnamed: 10':'ranking_2', 'Unnamed: 11':'ranking_3', 'Unnamed: 12':'ranking_4', 'Unnamed: 13':'ranking_5',
                                        'Unnamed: 14':'ranking_6'})
```

```python
star_wars.head()
```

## Most viewed movies

I'll use a bar graph to show the average movie for each of the rankings.

```python
%matplotlib inline

mean = star_wars[star_wars.columns[9:15]].mean()

mean.plot.bar()
plt.title('Average for each top 6 ranking')
plt.xlabel('Movie')
plt.ylabel('Average place')
```

As you can see in the above graph, the movie that most people seemed to like the best was _Star Wars: Episode V The Empire Strikes Back_ (ranking_5). The least favorite, on average, was _Star Wars: Episode III  Revenge of the Sith_ (ranking_3).

```python
star_wars[star_wars.columns[3:9]].sum()
```

```python
star_wars[star_wars.columns[3:9]].sum().plot.bar()
plt.ylabel('Number of people that have seen it')
plt.xlabel('Movie')
```

As expected, the "original" three movies were viewed more than the newer ones. This also correlates with being rated higher than the newer films.

## Splitting the data into groups

I'm now going to see if there are any interesting patterns when looking at males vs females, and people that are fans of the Star Trek franchise vs those who are not.

```python
males = star_wars[star_wars['Gender'] == 'Male']
females = star_wars[star_wars['Gender'] == 'Female']

# Sum of viewed movies and rankings by males and females
star_wars[star_wars.columns[2:]].groupby('Gender').sum()
```

```python
# Average male ranking for each movie
males[males.columns[9:15]].mean().plot.bar()
plt.title('Males')
```

```python
# average ranking for females for each movie
females[females.columns[9:15]].mean().plot.bar()
plt.title('Females')
```

```python
# average views for males for each movie
males[males.columns[3:9]].mean().plot.bar()
plt.title('Males')
```

```python
# average views for females for each movie
females[females.columns[3:9]].mean().plot.bar(title='Females')
```

Looking at the data above, it looks like both males and females prefered the "original" three movies to the newer ones, but the males liked them even more. Males have also seen all of the movies more than the females polled.
